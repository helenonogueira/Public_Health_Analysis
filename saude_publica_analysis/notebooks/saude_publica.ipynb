```python
def clean_data(df):
    # Excluir colunas irrelevantes
    df = df.drop(columns=['Country Code', 'Series Code'])

    df.rename(columns={'2017 [YR2017]': 'YR_2017', '2018 [YR2018]': 'YR_2018', '2019 [YR2019]': 'YR_2019', 
                       '2020 [YR2020]': 'YR_2020', '2021 [YR2021]': 'YR_2021', '2022 [YR2022]': 'YR_2022', 
                       '2023 [YR2023]': 'YR_2023'}, inplace=True)

    for year in range(2017, 2024):
        column_name = f'YR_{year}'
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')  # Converte e substitui NaN por não numéricos
        
    # Preencher valores faltantes com a média da coluna (apenas para colunas numéricas)
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        df[column].fillna(df[column].mean())



    
    return df  # Retorne o DataFrame modificado


# Carregue o DataFrame
df = pd.read_csv('C:\\Users\\Heleno\\Documents\\Projetos Portifólio\\saude_publica_analysis\\data\\World_Development_Indicators.csv')

# Limpe os dados e armazene o resultado em uma nova variável
df_cleaned = clean_data(df)

# Visualize as primeiras linhas da nova tabela
df_cleaned.head()
```
